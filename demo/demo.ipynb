```python
import matplotlib.pyplot as plt
%matplotlib inline
from mmseg.apis import init_model, inference_model, show_result_pyplot
from PIL import Image
```

#### Visualization

```python
config = 'mun_conf.py'
ckpt = 'mun_ckpt.pth'
device = 'cuda:0'
out = 'result.png'
opacity = 0.7
model = init_model(config, ckpt, device=device)
```

```python
img = 'images/ta5.jpg'
```

```python
result = inference_model(model, img)
vis_iamge = show_result_pyplot(model, img, result, out_file=out, opacity=opacity, show=False)
fig = plt.figure()
pred = Image.open('result.png').convert('RGB')
plt.imshow(pred)
plt.show()
```
